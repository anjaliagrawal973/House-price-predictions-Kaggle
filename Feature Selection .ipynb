```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

# for feature selection
from sklearn.linear_model import Lasso #to select the best feature
from sklearn.feature_selection import SelectFromModel

# to visualise al the columns in the dataframe
pd.pandas.set_option('display.max_columns', None)
```

```python
dataset=pd.read_csv('X_train.csv')
```

```python
dataset.head()
```

```python
y_train=dataset[['SalePrice']]
```

```python
# drop dependent feature from dataset
X_train=dataset.drop(['SalePrice'],axis=1)
```

```python
# alpha (equivalent of penalty).
# The bigger the alpha the less features that will be selected. 
#Then I use the selectFromModel object from sklearn, which will select the features which coefficients are non-zero

feature_sel_model = SelectFromModel(Lasso(alpha=0.000005, random_state=0)) # remember to set the seed/the random state value in this function
feature_sel_model.fit(X_train, y_train)
```

```python
feature_sel_model.get_support()
# True indicates that the feature is important and false otherwise
```

```python
# let's print the number of total and selected features

# this is how we can make a list of the selected or the important features
selected_feat = X_train.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format((X_train.shape[1])))
print('selected features: {}'.format(len(selected_feat)))

# coefficients that are not necessary have been shrank to 0
print('features with coefficients shrank to zero: {}'.format(
    np.sum(feature_sel_model.estimator_.coef_ == 0)))
```
